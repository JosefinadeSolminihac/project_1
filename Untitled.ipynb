```python
#Dependencies
import requests
import json
from config import api_key
```

```python
#Build the url
url=f"https://min-api.cryptocompare.com/data/top/mktcapfull?limit=10&tsym=USD&api_key={api_key}"
```

```python
#Get the information with json format, and isolates data items for easy reading
response=requests.get(url).json()
data=response["Data"]
data
```

```python
#Obtain the names of the top 10 main currencies considering Market Capitalization. Create a list and add the names to the list. 
indexes=range(0,10)
name_mktcap_list=[]
for index in indexes:
    name=data[index]["CoinInfo"]["FullName"]
    name_mktcap_list.append(name)
print(f"This are the top 10 cryptocurrencies in terms of market capitalization: {name_mktcap_list}")
```
